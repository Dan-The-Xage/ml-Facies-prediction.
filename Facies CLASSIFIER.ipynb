```python
from sklearn.model_selection import train_test_split    #importing the data from sklearn
X_train, X_test, y_train, y_test = train_test_split(predictor, target, test_size=0.3)  # splitting the  data into test and training data
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression #import Logistic regression algorith from sklearn
lrg = LogisticRegression().fit(X_train, y_train)      #training  
predicted_lrg = lrg.predict(X_test) # using the model to predict test data
predicted_lrg_trn = lrg.predict(X_train)   #
print('Training score: {:.2f}'.format(lrg.score(X_train,y_train))) #
print('Test score: {:.2f}'.format(lrg.score(X_test,y_test)))       #
from sklearn.metrics import confusion_matrix
conf_lrg = confusion_matrix(y_test, predicted_lrg)
print('Confusion_matrix for Logistic Regression' +'\n', conf_lrg)
lrg_acc = sklearn.metrics.accuracy_score(y_test, predicted_lrg)
print('Accuracy', lrg_acc)
```

```python
plt.subplot(1,2,1)
plt.title("Logistic regression model plot"     )
plt.scatter(y_train,predicted_lrg_trn)           #using plt.plot instead of plt.scatter
plt.legend(['y_train,predicted_lrg_trn'])      # to add a legend to the graph 



plt.subplot(1,2,2)
plt.title("         second logistic regression model plot")
plt.scatter(y_test,predicted_lrg)
plt.legend(['y_test,predicted_lrg'])
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
trc = DecisionTreeClassifier().fit(X_train, y_train)
predicted_trc = trc.predict(X_test)
predicted_trc_trn = trc.predict(X_train)
print('Training score: {:.2f}'.format(lrg.score(X_train,y_train))) #
print('Test score: {:.2f}'.format(lrg.score(X_test,y_test)))   
conf_trc = confusion_matrix(y_test, predicted_trc)
conf_trc
print('Confusion_matrix for Suport Decision' +'\n', conf_trc)
trc_acc = sklearn.metrics.accuracy_score(y_test, predicted_trc)
print('Accuracy', trc_acc)
```

```python
plt.subplot(1,2,1)
plt.title("Decision Tree model plot"     )
plt.scatter(y_train,predicted_trc_trn)           #using plt.plot instead of plt.scatter
plt.legend(['y_train,predicted_trc_trn'])      # to add a legend to the graph 



plt.subplot(1,2,2)
plt.title("         second Decision tree  model plot")
plt.scatter(y_test,predicted_trc)
plt.legend(['y_test,predicted_trc'])
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title("SVC model plot"     )
plt.scatter(y_train,predicted_svc_trn)           #using plt.plot instead of plt.scatter
plt.legend(['y_train,predicted_svc_trn'])      # to add a legend to the graph 



plt.subplot(1,2,2)
plt.title("         second SVC  model plot")
plt.scatter(y_test,predicted_svc)
plt.legend(['y_test,predicted_svc'])
plt.show()
```
